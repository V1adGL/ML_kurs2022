```python
# импорт pandas
import pandas as pd

# подгружаем исходники
data = pd.read_csv('train_dataset_train.csv')
data
```

```python
# избавляюсь от колонок, не имеющих значения в рамках задачи (дублирование данных и тп)
columns_to_drop = [
    'id',
    'ticket_id',
    'entrance_id',
    'station_id',
    'line_id',
    'entrance_nm'
]

data = data.drop(columns=columns_to_drop)
data
```

```python
# Для удобства поставил свои названия
data.rename(columns={'ticket_type_nm':'Тип билета',
                     'station_nm':'Название станции',
                     'pass_dttm':'Дата валидации',
                     'line_nm':'Название линии',
                     'time_to_under':'Время м/у 1 и 2 валидациями',
                    'label':'Станция выхода'},inplace=True)
data
```

```python
# Проверяю пробелы в данных
data.info()
```

```python
# поделю сутки на части
data['Дата валидации'] = pd.to_datetime(data['Дата валидации'])
data['Время, ч'] = data['Дата валидации'].dt.hour
data['Часть суток'] = data['Время, ч'].apply(
    lambda hour: 'утро' if hour>5 and hour<=11 else (
                 'день' if hour>11 and hour<=17 else (
                 'вечер' if hour>17 and hour<=0 else 'ночь')))
                
# выходной день
data['День недели'] = data['Дата валидации'].dt.weekday
data['Выходной']=data['День недели'].apply(
    lambda weekday: 1 if weekday>4 else 0)
columns_to_drop = [
    'Дата валидации',
    'Время, ч',
    'День недели',
]
data = data.drop(columns=columns_to_drop)

data
```

```python
# перевожу категориальные признаки в векторы

data = pd.get_dummies(data, prefix='Линия', columns=['Название линии'])



data
```

```python
from sklearn.feature_extraction import FeatureHasher

h_ticket = FeatureHasher(n_features=5, input_type='string')
hashed_Feature = h_ticket.fit_transform(data['Тип билета'])
hashed_Feature = hashed_Feature.toarray()
data = pd.concat([data, pd.DataFrame(hashed_Feature)], axis=1)

h_station = FeatureHasher(n_features=5, input_type='string')
hashed_Feature = h_station.fit_transform(data['Название станции'])
hashed_Feature = hashed_Feature.toarray()
data = pd.concat([data, pd.DataFrame(hashed_Feature)], axis=1)

h_daytime = FeatureHasher(n_features=5, input_type='string')
hashed_Feature = h_daytime.fit_transform(data['Часть суток'])
hashed_Feature = hashed_Feature.toarray()
data = pd.concat([data, pd.DataFrame(hashed_Feature)], axis=1)

data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Время м/у 1 и 2 валидациями', 'Станция выхода', 'Название станции', 'Тип билета','Часть суток'])
y = data['Время м/у 1 и 2 валидациями']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test, y_predict)
r2
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Время м/у 1 и 2 валидациями', 'Станция выхода', 'Название станции', 'Тип билета','Часть суток'])
y = data['Станция выхода']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model = LogisticRegression()
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import recall_score
recall=recall_score(y_test, y_predict, average='micro')
recall
```

```python
result = 0.5*r2 + 0.5*recall
result
```
